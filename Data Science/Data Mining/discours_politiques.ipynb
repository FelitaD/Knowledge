```python
import os
from requests_html import HTMLSession
session = HTMLSession()
```

```python
def get_speech(url):
    core_text = ''
    session = HTMLSession()
    row_web = session.get(url)
    # about = row_web.html.find('title') => <title>
    # for x in about :
    #     print(x.text)
    about = row_web.html.find('.discour--desc') # <class = discour--desc>
    save = False
    for x in about:
        if 'Texte intégral' in x.text:
            tmp = x.text.split('Texte intégral')
            core_text += ''.join(tmp[1:])
            save = True
        elif save:
            core_text += x.text
    # print(core_text)
    end_doc = core_text.lower().find('source http')
#    end_doc -= core_text[:38398][::-1].find(os.linesep) + 1
    core_text = core_text[:end_doc]
    end_doc = core_text.lower().find('http')
    core_text = core_text[:end_doc]
    return core_text[:end_doc]
```

```python
import nltk
import re
import string
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
doc_list = [get_speech(x['url']) for x in DONNEES_SOURCES]
```

```python
STOP_WORDS = stopwords.words('french')
custom_stop_words = ['le', 'de', 'et', 'être', 'un', 'qui', 'que', 'je', 'ce', 'notre', 'avoir', 'et', 'plus',
                    "s'est", "c'est", "j'ai"]
custom_stop_words.extend(STOP_WORDS)
```

```python
cleaned_doc = list()

for doc in doc_list:
    tmp = doc.split() # ['moi', 'president' ',' 'je' ]
    cleaned_doc.append(' '.join([x.lower() for x in tmp if x.lower() not in custom_stop_words ]))

cleaned_doc = [re.sub(r'\d+', '', x) for x in cleaned_doc]

ret = []
# remove pontuation
for doc in cleaned_doc:
    exclude = set(string.punctuation)
    ret.append(''.join([x for x in doc if x not in exclude]))
ret

# lemmatization avec spacy
import spacy
# python -m spacy download fr_core_news_md

def lemmatize(string):
    nlp = spacy.load('fr_core_news_md')
    # nlp => transforme une string en token (en gros, une liste de mots)
    doc = nlp(string)
    # jointure pour avoir un text continu
    return ' '.join([x.lemma_ for x in doc])

cleaned_doc =  [ lemmatize(x) for x in ret]
```
