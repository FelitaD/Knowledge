# Travelling salesman

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(123)
n = 20
pts = np.random.rand(n, 2)
```

```python
pts
```

```python
def show_trip(pts):
    x = np.append(pts[:, 0], pts[0, 0])
    y = np.append(pts[:, 1], pts[0, 1])
    plt.plot(x, y)
    plt.scatter(x, y)
    
    for i in range(len(pts)):
        plt.annotate(i, pts[i])
```

```python
x = np.append(pts[:, 0], pts[0, 0])
x
```

```python
show_trip(pts)
```

```python
def dist(pts):
    d = np.sqrt(np.sum((pts[0] - pts[-1]) ** 2))
    
    for i in range(len(pts) - 1):
        d += np.sqrt(np.sum((pts[i + 1] - pts[i]) ** 2))
        
    return d
```

```python
dist(pts)
```

```python
def dist(pts):
    return np.sum(np.linalg.norm(pts - np.roll(pts, -2), axis=1))
```
